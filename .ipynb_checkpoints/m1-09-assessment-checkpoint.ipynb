```python
# Ard-arda hər şəhərdə neçə gün NaN dəyərlər buraxılıb:
# buna baxmaq lazımdır ki interpolasiya edib-edə bilməyəciyimizi anlayaq
```

```python
import pandas as pd

def max_consecutive_nans(series):
    # Convert True for NaN, False otherwise
    is_nan = series.isna()
    # Group by consecutive non-NaNs
    return is_nan.groupby((is_nan != is_nan.shift()).cumsum()).sum().max()

# Apply the function for each city
max_gaps = df.groupby('city')['pm25'].apply(max_consecutive_nans)

max_gaps
```

PM2.5 — havadakı çox xırda toz zərrəciklərinin miqdarıdır.

PM sütunundakı NaN dəyərləri doldurmaq üçün mən interpolation metodundan istifadə edəcəyəm.

**Xətti İnterpolyasiya (Linear Interpolation)** çox sadə bir məntiqə əsaslanır: 

əgər əlimizdə ardıcıl günlərin məlumatı varsa və ortada bir gün çatışmırsa, həmin günü qonşu günlərin "ortalaması" ilə birləşdiririk.Bunu daha rahat başa düşmək üçün belə bir misala baxaq:

      -Bazar ertəsi: Hava çirkliliyi (PM_{2.5}) 20 vahid olub.
      
      -Çərşənbə axşamı: Məlumat yoxdur (NaN).
      
      -Çərşənbə: Hava çirkliliyi 30 vahid olub.
      
 İnterpolyasiya nə edir?
 
O fərz edir ki, hava çirkliliyi birdən-birə yox, tədricən dəyişir. Bazar ertəsindən (20) Çərşənbəyə (30) doğru bir düz xətt çəkir. Bu xəttin tam ortası Çərşənbə axşamına düşür və proqram həmin boşluğa 25 rəqəmini yazır.

Niyə bu üsul ən yaxşısıdır?
   
   Hava temperaturu və ya çirkliliyi adətən saniyələr içində kəskin dəyişmir. Əgər dünən soyuq idisə və sabah da soyuq olacaqsa, bu günün də soyuq olduğunu təxmin etmək "Xətti İnterpolyasiya"dır.
   Şəhər fərqi: Biz bu kodu groupby('city') ilə yazırıq ki, Londonun havası ilə Madridin havası bir-birinə qarışmasın. Hər şəhərin boşluğunu öz günlərinə əsasən doldurur.

İnterpolyasiya (interpolation) həm mean (orta rəqəm), həm də median-dan çox daha məntiqlidir.Niyə belə?

 --- 1. Zaman Ardıcıllığı (Time-Series):
     
     Hava çirkliliyi və temperatur kimi məlumatlar "zaman ardıcıllığı"dır. Əgər bütün Madrid şəhəri üzrə 6 aylıq mean (orta qiymət) tapsanız, bu, yanvar ayındakı soyuq günlərdə də, iyun ayındakı isti günlərdə də eyni rəqəmi yazacaq. Amma interpolyasiya dünənki və sabahkı rəqəmə baxdığı üçün həmin günün real trendinə uyğun rəqəm qoyur.
     
 --- 2. Sıçrayışlardan qorunma:
     
     Məsələn, bir şəhərdə bir gün fabrik qəzası olub və hava çirkliliyi $PM_{2.5}$ 200-ə qalxıb (outlier).Mean istifadə etsəniz, bu 200 rəqəmi ortalamanı yuxarı çəkəcək və digər sakit günlərdəki boşluqlara süni şəkildə yüksək rəqəmlər yazacaq.İnterpolyasiya isə yalnız həmin günün ətrafına baxır, uzaqdakı kəskin artımlardan təsirlənmir.

**Bəs medianı harda istifadə edirik?**

Əgər boşluq məlumatın tam ən başında (yanvarın 1-i) və ya tam ən sonunda (iyunun sonu) olarsa, interpolyasiyanın "birləşdirməyə" nöqtəsi olmur. Yalnız həmin nadir hallarda şəhərin median rəqəmindən istifadə olunur.

**Onda saniyələr içində kəskin dəyişən dəyərlər üçün mean,median ama tədricən dəyişən dəyərlər üçün interpolasiya daha yaxşıdır?**

Bəli. Amma burada kiçik bir incəlik var: zaman faktoru.

Bu iki vəziyyəti müqayisə edək:

1. **Tədricən dəyişən dəyərlər (Hava temperaturu, PM2.5):**


   Bunlar təbiət hadisələridir və bir-birinə zəncirvari bağlıdırlar. Saat 12:00-da hava $20^\circ C$ dərəcədirsə, saat 13:00-da birdən-birə $-10^\circ C$ ola bilməz. Arada mütləq keçid olmalıdır.
   
   -Niyə İnterpolyasiya?

   Çünki o, bu "keçid" məntiqini (məsələn: 20, 21, 22...) bərpa edir.

2. **Kəskin və ya qeyri-müntəzəm dəyişən dəyərlər**

   (Məsələn: Səhmlərin qiyməti, mağaza satışları). Təsəvvür edin ki, bir mağazada bazar ertəsi 100 çörək satılıb, çərşənbə isə 500. Çərşənbə axşamı (boşluq olan gün) mütləq 300 satıldığını deyə bilmərik. Bəlkə o gün futbol oyunu olub və satışlar kəskin artıb/azalıb?

   -Niyə Mean/Median? Belə hallarda trendi təxmin etmək risklidir. Ona görə də həmin ayın və ya həmin günlərin "orta rəqəmini" (Mean) qoymaq daha təhlükəsizdir ki, ümumi statistikaya çox zərər dəyməsin.

```python
# Filling pm25
df['pm25'] = df.groupby('city', group_keys=False)['pm25'].apply(
    lambda x: x.interpolate(method='linear').fillna(x.median())
)
# niyə fillna(x.median()) yazmışıq? --- 
# İnterpolyasiya işləmək üçün mütləq boşluqdan həm əvvəl, həm də sonra bir rəqəm görməlidir ki, onların arasını birləşdirsin.
# Amma təsəvvür edin ki:
# Məlumatın tam başlanğıcı: 1 yanvar tarixində məlumat yoxdur (NaN). Yanvarın 1-dən əvvəl rəqəm olmadığı üçün interpolyasiya "xətt çəkməyə" nöqtə tapmır və oranı boş (NaN) saxlayır.
# Məlumatın tam sonu: 30 iyun tarixində məlumat yoxdur. İyunun 30-dan sonra məlumat olmadığı üçün interpolyasiya yenə çarəsiz qalır.
# Məhz belə hallarda .fillna(x.median()) işə düşür. O deyir ki: "Əgər interpolyasiya boşluğu doldura bilmədisə (yəni boşluq cədvəlin ən başındadırsa və ya sonundadırsa), onda həmin şəhərin orta (median) göstəricisini ora yaz."

# CHECK: Confirm everything is filled
print("\n--- Missing Values After Cleaning ---")
print(df.isnull().sum())
```

#### **group_keys = False nəyi bildirir?**

* group_keys=False olanda: Nəticə yalnız date indeksi ilə qayıdır. Pandas görür ki, bu yeni sütun köhnə cədvəlin "ölçüsünə" tam uyğundur və onu rahatca yerinə yapışdırır.

* group_keys=True olanda: Nəticədə həm city, həm date indeksi olur. Siz bu yeni "ikiqat indeksli" sütunu köhnə "tək indeksli" cədvələ daxil etmək istəyəndə, Pandas çaşır və ya xəta verir, ya da bütün rəqəmləri NaN edir (çünki indekslər bir-birinə "tıqqıltı" ilə oturmur).

```python
# 8.Compare volatility between two cities using a metric you define and justify.
```

Volatility maliyyə və statistika terminidir. Sadə dildə desək: "Rəqəmlər nə dərəcədə qeyri-sabitdir?"

Bir şəhərdə hava çirkliliyi bir gün 10, o biri gün 100, növbəti gün yenə 15 olursa, bu şəhər yüksək volatilliyə (dəyişkənliyə) malikdir. Əgər rəqəmlər həmişə 40-50 arası qalırsa, bu stabil (aşağı volatilli) şəhərdir.

Tapşırıqda nə istəyir?
Sizə deyir ki, iki şəhər seçin (məsələn, London və Berlin) və hansının havasının daha "oynaq" (qeyri-sabit) olduğunu sübut edin. Bunun üçün bir metrika (ölçü vahidi) təyin etməlisiniz.

```python
# 1. Şəhərlər üzrə Standart Meyllənməni (Volatility) hesablayırıq
volatility = df.groupby('city')['pm25'].std()

# 2. İki şəhəri müqayisə edirik (Məsələn London və Madrid)
city1 = 'London'
city2 = 'Madrid'

v1 = volatility[city1]
v2 = volatility[city2]

print(f"{city1} Volatility: {v1:.2f}")
print(f"{city2} Volatility: {v2:.2f}")

# 3. Hansı daha böyükdürsə, o daha dəyişkəndir
if v1 > v2:
    print(f"Nəticə: {city1} daha dəyişkən (volatile) havaya malikdir.")
else:
    print(f"Nəticə: {city2} daha dəyişkən (volatile) havaya malikdir.")
```

```python
#  9.Create a month-by-city pivot table of average pm25.
```

```python
df.index = pd.to_datetime(df.index)
df.index
```

```python
# Ay adlarını çıxarıb yeni sütuna qoyuruq
df['month'] = df.index.month_name()
# .month_name() - ay adlarını bütün date-dən extract edir
```

```python
# Pivot table-ı yaradırıq
pivot_result = df.pivot_table(index='month', 
                              columns='city', 
                              values='pm25', 
                              aggfunc='mean')
```

```python
pivot_result
```

```python
# 4. Ayların təqvim sırasını təyin edirik
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']

# 5. Sıralamanı düzəldib nəticəni göstəririk
pivot_result = pivot_result.reindex(month_order).dropna(how='all')
pivot_result
```

* dropna(how='all') :

  "Əgər bir ayın (sətrin) bütün şəhərlər üzrə dəyərləri boşdursa, həmin ayı cədvəldən tamamilə çıxart".

* how='all' nə deməkdir?

   - how='all': Yalnız sətirdəki bütün sütunlar boşdursa silir. (Məsələn, İyul ayında heç bir şəhərdə ölçmə yoxdursa, İyul silinir).

   - how='any': Əgər sətirdə hətta bir dənə boş xana varsa, bütün sətri silir. (Bunu istifadə etmə, yoxsa bir şəhərin məlumatı çatmayanda bütün ayı silər).
